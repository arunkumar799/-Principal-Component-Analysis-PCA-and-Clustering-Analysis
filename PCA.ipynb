# Comparison and Analysis

```python
# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 components for visualization
pca_data = pca.fit_transform(new_features)

# Fit KMeans on original data
kmeans_orig = KMeans(n_clusters=3).fit(new_features)
labels_orig = kmeans_orig.labels_

# Fit KMeans on PCA-transformed data
kmeans_pca = KMeans(n_clusters=3).fit(pca_data)
labels_pca = kmeans_pca.labels_

# Silhouette scores
silhouette_orig = silhouette_score(new_features, labels_orig)
silhouette_pca = silhouette_score(pca_data, labels_pca)
print(f"Silhouette Score on Original Data: {silhouette_orig}")
print(f"Silhouette Score on PCA Data: {silhouette_pca}")

# Plotting clusters for comparison
plt.figure(figsize=(12, 6))

# Original Data clusters
plt.subplot(1, 2, 1)
plt.scatter(features1['Alcohol'], features1['Magnesium'], c=labels_orig, cmap='viridis')
plt.title('Clustering on Original Data')
```

### After applying the PCA the resultant features showned in the graph

```python
plt.subplot(1, 2, 1)
plt.scatter(New_features['Alcohol'], New_features['Magnesium'], c=labels_orig, cmap='viridis')
plt.title("PCA-transformed data clustering results")
```
